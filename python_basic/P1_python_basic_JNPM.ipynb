##P1T6_Python_Basic_JNPM

Programa en python que genera un listado con las edades de n habitantes de una ciudad, donde la cantidad de habitantes sea un valor aleatorio entre 1000 y 5000 y la edad de cada persona también debe ser un valor aleatorio entre 1 y 100 años. Al final debe indicar cual es la población total y el porcentaje de menores y mayores de edad



```
ages = [randint(1, 100) for i in range(randint(1000, 5000))]
adults = round((len([age for age in ages if age >= 18]) * 100) / len(ages), 2)
print(f"Población total: {len(ages)}, mayores de edad: {adults} %, menores de edad: {round(100 - adults, 2)} %")
```

##P1T7_Python_Basic_JNPM

Programa en python que determina el candidato ganador de 6 candidatos que se disputan la alcaldía. Los votos de cada candidato son generados de forma aleatoria entre 1 y 25000 votos máximo.

```
resultado = np.array([[i + 1, randint(1, 25000)] for i in range(6)])
ganador = np.where(resultado[:, 1] == resultado[:, 1].max())
print(f"El ganador es el candidato número {int(resultado[ganador, 0])} {int(resultado[ganador, 1])} votos")

# Gráficas de datos 2D
diagrama = plt.figure('Gráfica de barras')
eje = diagrama.add_subplot(1, 1, 1)
nombres = ['1 – socio-democrático', '2 – Ultra-derechista', '3 – primero los ricos', '4 –Centro demoniaco',
           '5 – Cambio invertido', '6 – Alianza queremos más pobres']
eje.bar(range(len(resultado)), resultado[:, 1])
eje.set_xticks(range(len(resultado[:, 1])))
eje.set_xticklabels(nombres, rotation='vertical')
plt.ylabel("Total votos")
plt.xlabel("Candidatos")
plt.show()
```

##P1T8_Python_Basic_JNPM

Programa en python donde 6 candidatos se disputan la gobernación del departamento. Los votos de cada candidato son
generados de forma aleatoria entre 1 y 4500 votos máximo para cada uno de los 123 municipios del departamento, el
software debe determinar quien fue el ganador.

**Nota**: Creamos una matriz llamada votos, compuesta por 123 filas y 6 columnas cuyo valor de cada celda sea un número
aleatorio de 1 a 4500

```
resultado = np.random.randint(4500, size=(123, 6))
suma_mpio = np.array([sum(resultado[:, i]) for i in range(6)])
ganador = np.where(suma_mpio[:] == suma_mpio[:].max())
print(f"El ganador es el candidato número {int(ganador[0]) + 1} con {int(suma_mpio[ganador])} votos")
```

##P1T9_Python_Basic_JNPM

Programa en python que simula una carrera de la fórmula uno, donde existen 5 vehículos:



*   Renault
*   Ferrari
*   Mercedes
*   Ford
*   Red bull

La carrera está diseñada para 4 vueltas donde el tiempo mínimo es 8.2 segundos y el máximo es 10 segundos.

Diseñe la matriz 5x4 donde determine cual es el vehículo ganador donde los tiempos de cada vehículo por vuelta sea un
número aleatorio.

```
resultado = np.random.uniform(8.2, 10.1, (5, 4))
tiempo_individual = np.array([sum(resultado[i, :]) for i in range(5)])
ganador = np.where(tiempo_individual[:] == tiempo_individual[:].min())
print(f"El ganador fue el vehículo {int(ganador[0]) + 1} con {round(float(tiempo_individual[ganador]), 4)} segundos")

# Gráfico de dispersión
eje_x = [1, 2, 3, 4, 5]
eje_y = tiempo_individual[:]
diagram = plt.figure()
diagram.add_subplot(111)
plt.xlabel('Pilotos')
plt.ylabel('Tiempo total')
plt.scatter(eje_x, eje_y, c=eje_x)
plt.show()
```
